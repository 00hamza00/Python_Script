```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandasql import sqldf
```

```python
filepath = '/Users/Hamza/Downloads/Python Project/'

dfpopulation = pd.read_csv(filepath + 'population.csv')
```

```python
update_df = dfpopulation.drop(dfpopulation.index[0:7])
update_df.columns = ['Year','Population']
update_df.set_index('Year',inplace=True)
update_df.head()
```

```python
def pysqldf(query):
    return sqldf(query, globals())
```

```python
last10y_query = "Select * from update_df order by Year asc limit 10;"
last5y_query = "Select * from update_df order by Year desc limit 5;"
selecting_year_inrange = "Select * from update_df where Year BETWEEN 2000 and 2015;"
```

```python
y10 = pysqldf(last10y_query)
y5 = pysqldf(last5y_query)
yrange = pysqldf(selecting_year_inrange)
```

```python
fig = plt.figure(figsize=(10,10))
fig.add_subplot(1,1,1)
x = "Year"
y= "Population"

sns.set_theme(style="ticks")
sns.scatterplot( data=yrange, x=x,y=y)


```
